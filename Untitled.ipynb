11. Property Sub Type: Further categorization of the property type.
12. Number of Rooms: The number of rooms in the property.
13. Usage: The designated usage of the property (e.g., residential, commercial, industrial).
14. Nearest Metro: The nearest metro station to the property.
15. Nearest Mall: The nearest shopping mall to the property.
16. Nearest Landmark: The closest landmark or point of reference to the property.
17. Parking: Information about parking facilities.
18. No of Units: The number of units (e.g., apartments, villas) within the property.
19. Master Project: The master project or development name.
20. Project: The project name or specific subdivision within the master project.
